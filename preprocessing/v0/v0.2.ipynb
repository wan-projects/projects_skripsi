```python
import os

folder_path = '../Oprek/Ori_Hanacaraka'
kelas_folders = [f for f in os.listdir(folder_path) if os.path.isdir(os.path.join(folder_path, f))]

jumlah_data_per_kelas = {}

for kelas_folder in kelas_folders:
    kelas_folder_path = os.path.join(folder_path, kelas_folder)
    image_files = [f for f in os.listdir(kelas_folder_path) if f.endswith('.png')]
    
    jumlah_data_per_kelas[kelas_folder] = len(image_files)

# Menampilkan jumlah data per kelas
for kelas, jumlah_data in jumlah_data_per_kelas.items():
    print(f"Kelas: {kelas}, Jumlah Data: {jumlah_data}")
```

```python
# Visualisasi data (Opsional)
import matplotlib.pyplot as plt

plt.bar(jumlah_data_per_kelas.keys(), jumlah_data_per_kelas.values())
plt.xlabel('Kelas')
plt.ylabel('Jumlah Data')
plt.title('Jumlah Data per Kelas')
plt.show()
```
