```python
import os
import cv2

def cek_label_gambar(gambar_path, dataset_root):
    # Mendapatkan nama gambar dari path
    nama_gambar = gambar_path.split("/")[-1]

    # Mendapatkan nama subdirektori (kelas) dari path
    kelas = gambar_path.split("/")[-2]

    # Membaca gambar menggunakan OpenCV
    gambar = cv2.imread(gambar_path)

    # Memeriksa apakah nama gambar ada dalam subdirektori (kelas)
    if os.path.exists(os.path.join(dataset_root, kelas, nama_gambar)):
        print(f"Gambar '{nama_gambar}' dalam kelas '{kelas}' sudah ada")
    else:
        print(f"Gambar '{nama_gambar}' dalam kelas '{kelas}' belum ada")

# Contoh penggunaan
dataset_root = '../Oprek/Ori_Hanacaraka'  # Ganti dengan path root dari dataset Anda

# Iterasi melalui setiap subdirektori (kelas)
for kelas in os.listdir(dataset_root):
    kelas_path = os.path.join(dataset_root, kelas)

    # Memeriksa setiap gambar dalam subdirektori (kelas)
    for nama_gambar in os.listdir(kelas_path):
        gambar_path = os.path.join(kelas_path, nama_gambar)
        cek_label_gambar(gambar_path, dataset_root)
```
